```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
school_num = len(school_data)

#Calculate the total number of students
student_num=len(student_data)

#Calculate the total budget
tot_budget = school_data['budget'].sum()

#Calculate the average math score
math_av = school_data_complete['math_score'].sum()/student_num

#Calculate the average reading score
reading_av = school_data_complete['reading_score'].sum()/student_num

#Calculate the percentage of students with a passing math score (70 or greater)
school_data_complete.loc[school_data_complete['math_score']>=70, 'math_pass'] = True
percent_math_pass = school_data_complete.groupby('math_pass')['math_pass'].count()/student_num * 100

#Calculate the percentage of students with a passing reading score (70 or greater)
school_data_complete.loc[school_data_complete['reading_score']>=70, 'reading_pass'] = True
percent_reading_pass = school_data_complete.groupby('reading_pass')['reading_pass'].count()/student_num * 100

#Calculate the percentage of students who passed math and reading (% Overall Passing)
school_data_complete.loc[school_data_complete['reading_pass'] & school_data_complete['math_pass'], 'overall_pass'] = True
percent_passing = school_data_complete.groupby('overall_pass')['overall_pass'].count()/student_num*100
```

```python
#Create a dataframe to hold the above results
district_df = pd.DataFrame({'Total Schools': school_num, 
                            'Total Students': student_num,
                            'Total Budget': tot_budget,
                            'Average Math Score': math_av,
                            'Average Reading Score': reading_av,
                            '% Passing Math': percent_math_pass,
                            '% Passing Reading': percent_reading_pass,
                            '% Overall Passing': percent_passing})

#Optional: give the displayed data cleaner formatting
district_df['Total Students'] = district_df['Total Students'].map("{:,}".format)
district_df['Total Budget'] = district_df['Total Budget'].map("${:,.2f}".format)
district_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Create an overview table that summarizes key metrics about each school, including:
#School Name,School Type,Total Students,Total School Budget
overview_df = pd.DataFrame({"School Name": school_data["school_name"],
                            "School Type": school_data["type"],
                            "Total Students": school_data["size"],
                            "Total School Budget": school_data["budget"]
                           })
overview_df = overview_df.sort_values(by = "School Name")
overview_df = overview_df.set_index('School Name')

#Per Student Budget
overview_df["Per Student Budget"] = overview_df["Total School Budget"]/overview_df["Total Students"]

#Average Math Score
school_grouped_mean_df = school_data_complete.groupby(["school_name"]).mean()
overview_df["Average Math Score"] = school_grouped_mean_df["math_score"]

#Average Reading Score
overview_df["Average Reading Score"] = school_grouped_mean_df["reading_score"]

#% Passing Math
school_grouped_count_df = school_data_complete.groupby(["school_name"]).count()
overview_df["% Passing Math"] = school_grouped_count_df["math_pass"]/school_grouped_count_df["math_score"]*100

#% Passing Reading
overview_df["% Passing Reading"] = school_grouped_count_df["reading_pass"]/school_grouped_count_df["reading_score"]*100

#% Overall Passing (The percentage of students that passed math and reading.)
overview_df["% Overall Passing"] = school_grouped_count_df["overall_pass"]/school_grouped_count_df["reading_score"]*100

#make it pretty
overview_df['Total Students'] = overview_df['Total Students'].map("{:,}".format)
overview_df['Total School Budget'] = overview_df['Total School Budget'].map("${:,.2f}".format)
overview_df['Per Student Budget'] = overview_df['Per Student Budget'].map("${:,.2f}".format)
overview_df
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
top_schools = overview_df.sort_values(by='% Overall Passing', ascending = False)
top_schools.head()
```

## Bottom Performing Schools (By % Overall Passing)

* Sort and display the five worst-performing schools by % overall passing.

```python
bottom_schools = overview_df.sort_values(by='% Overall Passing')
bottom_schools.head()
```
